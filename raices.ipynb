# Determinar las raices del siguiente polinomio

    x5−6∗x4+2∗x3+20∗x2−27∗x+10=0.

```python
# Con el metodo de newton

import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize

def polinomio(x):
    return x**5 - 6*x**4 + 2*x**3 + 20*x**2 - 27*x + 10

array = np.arange(-50,50,10)

resultado = [optimize.newton(polinomio,x) for x in array]

print(f'Con el metodo de newton tenemos el siguiente resultado {resultado}')


analytical_result = np.sign(array) * np.abs(array)**(1/3)
fig, ax = plt.subplots()
ax.plot(array, analytical_result, 'o')
ax.plot(array, resultado, '.')
ax.set_xlabel('$a$')
ax.set_ylabel('x5−6∗x4+2∗x3+20∗x2−27∗x+10=0')
plt.show()

" El codigo fue hecho solo leyendo documentacion, intente entender todo"

# Con el metodo de bisect

resultado_bistec = optimize.bisect(polinomio, -20,20)

print("La resultado es:", resultado_bistec)


```
